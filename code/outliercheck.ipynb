```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

OUTLIERS FOR players_teams

```python
df=pd.read_csv("C:\\Users\\L\\Downloads\\New folder\\players_teams.csv")
num_col = df.select_dtypes(include=['number'])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

sns.boxplot(data=num_col, orient='v', width=0.5, ax=axes[0])

axes[0].set_xlabel('Columns')
axes[0].set_ylabel('Original Y-Axis Label')
axes[0].set_title('Original Graph')

axes[0].set_xticks(range(len(num_col.columns))) 
axes[0].set_xticklabels(num_col.columns, rotation=90)

axes[0].set_ylim(num_col.values.min()-50, num_col.values.max()+50)

axes[0].set_title('players_teamsAll plot')

sns.boxplot(data=num_col, orient='v', width=0.5, ax=axes[1])

axes[1].set_xlabel('Columns')
axes[1].set_ylabel('Original Y-Axis Label')
axes[1].set_title('Original Graph')

axes[1].set_xticks(range(len(num_col.columns))) 
axes[1].set_xticklabels(num_col.columns, rotation=90)

axes[1].set_ylim(0, 20)

axes[1].set_title('players_teams Attributes y, 0 to 20');
```
